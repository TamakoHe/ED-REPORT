```python
import numpy as np 
import plotly.graph_objects as go  
```

```python
def graph(file:str, label="V2"):
    def get_v2(text:str):
        for idx,t in enumerate(text):
            if t=='V':
                sub_txt=""
                for sub_t in text[idx+3:]:
                    if sub_t!="(":
                        sub_txt+=sub_t
                    else:
                        break
                if sub_txt.find("m")!=-1:
                    return 1e-3*float(sub_txt.replace("m",""))
                return float(sub_txt)    
    data={}
    with open(file) as text:
        lines = text.readlines()
        for line in lines:
            if line.startswith("Step"):
                this_v2=get_v2(line)
                data[this_v2]={}
            else:
                i,v=line.split("\t")
                i=float(i)
                v=float(v)
                data[this_v2][v]=i 
    fig=go.Figure()
    for v2 in data.keys():
        fig.add_trace(go.Scatter(y=list(data[v2].keys()),x=list(data[v2].values()),name=f"{label}={v2}"))
    fig.update_layout(xaxis_title="Voltage(V)", yaxis_title="Current(A)")
    fig.show()
    return data
```

```python
graph("task1.txt")
data2=graph("task2.txt","VGS")
```

```python
def graph3(data, sel=[-1.1,-1.6]):
    fig=go.Figure()
    for v2 in data.keys():
        if v2 in sel:
            fig.add_trace(go.Scatter(y=list(data[v2].keys()),x=list(data[v2].values()),name=f"V2={f"VGS={v2}"}"))
    fig.show()
graph3(data2)
```
