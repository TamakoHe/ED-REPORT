```python
import numpy as np
```

```python
def get_data_from_text(file_path:str):
    f=open(file_path,'r')
    lines=f.readlines()
    Vins=[]
    Vbes=[]
    Ibs=[]
    Ics=[]
    Ies=[]
    betas=[]
    Vccs=[]
    Vces=[]
    for idx, line_text in enumerate(lines):
        if idx==0:
            continue
        Vin, Vcc, Vce,Vbe,Ib, Ic=line_text.split('\t')
        Vins.append(float(Vin))
        Vccs.append(float(Vcc))
        Vbes.append(float(Vbe))
        Ibs.append(float(Ib))
        Ics.append(float(Ic))
        Ies.append(float(Ic)+float(Ib))
        betas.append(float(Ic)/float(Ib))
        Vces.append(float(Vce))
    return {"Vin":Vins, "Vcc":Vccs, "Vbe":Vbes,"Ib":Ibs, "Ic":Ics, "Ie":Ies, "beta":betas, "Vce":Vces}
```

```python
data_1=get_data_from_text("1.txt")
data_2=get_data_from_text("2.txt")
data_3=get_data_from_text("3.txt")
data_4=get_data_from_text("4.txt")
data_5=get_data_from_text("5.txt")
data_6=get_data_from_text("6.txt")
data_8=get_data_from_text("8.txt")
data_10=get_data_from_text("10.txt")
data_15=get_data_from_text("15.txt")
```

```python
import plotly.graph_objects as go 
```

```python
def draw_Vbe_Ic(data:dict, vin=15):
    fig=go.Figure()
    fig.add_trace(go.Scatter(x=data["Vbe"], y=data["Ic"], mode="lines+markers", name="Vbe_Ic"))
    fig.update_layout(title=f"Vbe_Ic vin={vin}", xaxis_title="Vbe", yaxis_title="Ic")
    fig.show()
def draw_Vce_Ic(data:dict, vin=15):
    fig=go.Figure()
    fig.add_trace(go.Scatter(x=data["Vce"], y=data["Ic"], mode="lines+markers", name="Vbe_Ic"))
    fig.update_layout(title=f"Vce_Ic vin={vin}", xaxis_title="Vce", yaxis_title="Ic")
    fig.show()
```

```python
draw_Vbe_Ic(data_15,15)
draw_Vce_Ic(data_15,15)
draw_Vbe_Ic(data_10,10)
draw_Vce_Ic(data_10,10)
draw_Vbe_Ic(data_8,8)
draw_Vce_Ic(data_8,8)
draw_Vbe_Ic(data_6,6)
draw_Vce_Ic(data_6,6)
draw_Vbe_Ic(data_5,5)
draw_Vce_Ic(data_5,5)
draw_Vbe_Ic(data_4,4)
draw_Vce_Ic(data_4,4)
draw_Vbe_Ic(data_3,3)
draw_Vce_Ic(data_3,3)
draw_Vbe_Ic(data_2,2)
draw_Vce_Ic(data_2,2)
draw_Vbe_Ic(data_1,1)
draw_Vce_Ic(data_1,1)
```
